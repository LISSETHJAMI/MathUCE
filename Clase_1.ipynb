## Fundamentos de Lógica y Conjuntos

### 1. Lógica Proposicional
#### Conectores Lógicos:
- **Conjunción (∧):**
  - Representa "y".
  - Verdadero solo si ambas proposiciones son verdaderas.
  - Ejemplo: `p ∧ q`

- **Disyunción (∨):**
  - Representa "o".
  - Verdadero si al menos una proposición es verdadera.
  - Ejemplo: `p ∨ q`

- **Negación (¬):**
  - Representa "no".
  - Invierte el valor de verdad de una proposición.
  - Ejemplo: `¬p`

- **Implicación (→):**
  - Representa "si... entonces".
  - Falsa solo si la primera proposición es verdadera y la segunda es falsa.
  - Ejemplo: `p → q`

- **Doble Implicación (↔):**
  - Representa "si y solo si".
  - Verdadera si ambas proposiciones tienen el mismo valor de verdad.
  - Ejemplo: `p ↔ q`

#### Proposiciones:
- Ejemplos:
  - `p`: "Está lloviendo."
  - `q`: "Voy a llevar un paraguas."

#### Tablas de Verdad:
- Herramienta para analizar todas las posibles combinaciones de verdad de las proposiciones.
- Ejemplo para la conjunción (`p ∧ q`):



```
# | `p` | `q` | `p ∧ q` |
  |----|----|--------|
  | V  | V  | V      |
  | V  | F  | F      |
  | F  | V  | F      |
  | F  | F  | F      |

```

#### Cuantificadores:
- **Cuantificador Existencial (∃):**
  - "Existe al menos uno".
  - Ejemplo: `∃x ∈ ℝ` tal que `x² = 1`

- **Cuantificador Universal (∀):**
  - "Para todo".
  - Ejemplo: `∀x ∈ ℝ, x² ≥ 0`

### 2. Teoría de Conjuntos

#### Definición:
- Un conjunto es una colección bien definida de objetos.
- Ejemplos: `A = {1, 2, 3}`

#### Notación:
- **Elementos:** `a, b, c`
- **Conjuntos:** `A, B, C`
- **Pertenencia:** `a ∈ A`
- **No pertenencia:** `b ∉ A`

#### Tipos de Conjuntos:
- **Conjunto Vacío (∅):**
  - No contiene elementos.
  - Notación: `∅`

- **Conjunto Universal (U):**
  - Contiene todos los elementos bajo consideración.
  - Notación: `U`

#### Operaciones con Conjuntos:
- **Intersección (∩):**
  - Elementos comunes a ambos conjuntos.
  - Notación: `A ∩ B`

- **Unión (∪):**
  - Todos los elementos presentes en al menos uno de los conjuntos.
  - Notación: `A ∪ B`

- **Diferencia (A - B):**
  - Elementos que están en `A` pero no en `B`.
  - Notación: `A - B`

- **Complemento (A^c):**
  - Elementos que no están en `A`.
  - Notación: `A^c`

#### Diagramas de Venn:
- Representación gráfica de las relaciones entre conjuntos.

#### Propiedades de las Operaciones:
- **Conmutativa:**
  - `A ∪ B = B ∪ A`
  - `A ∩ B = B ∩ A`

- **Asociativa:**
  - `(A ∪ B) ∪ C = A ∪ (B ∪ C)`
  - `(A ∩ B) ∩ C = A ∩ (B ∩ C)`

- **Distributiva:**
  - `A ∪ (B ∩ C) = (A ∪ B) ∩ (A ∪ C)`
  - `A ∩ (B ∪ C) = (A ∩ B) ∩ (A ∩ C)`

### 3. Ejemplos y Ejercicios

#### Ejemplo 1:
- Dados los conjuntos `A = {1, 2, 3}` y `B = {2, 3, 4}`:
  - Intersección: `A ∩ B = {2, 3}`
  - Unión: `A ∪ B = {1, 2, 3, 4}`
  - Diferencia: `A - B = {1}`

#### Ejemplo 2:
- En lógica proposicional, si `p` es "Llueve" y `q` es "Llevo un paraguas":
  - Implicación: "Si llueve, entonces llevo un paraguas" se escribe como `p → q`.


## 2. Teoría de Conjuntos

### Definición:
- Un conjunto es una colección de elementos.
- Ejemplos: números, personas, formas geométricas, etc.


### Notación:
- **Mayúsculas:** Conjuntos (A, B, Ω).
- **Minúsculas:** Elementos (x, y).
- **∈:** Pertenencia (x ∈ A).
- **∉:** No pertenencia (x ∉ A).


### Relaciones entre Conjuntos:
- **Subconjunto:** A ⊆ B (A está contenido en B).
- **Subconjunto estricto:** A ⊂ B (A está estrictamente contenido en B).

### Operaciones entre Conjuntos:
- **Intersección (∩):** Elementos comunes entre conjuntos.
  - Notación: x ∈ A ∩ B → x ∈ A ∧ x ∈ B

- **Unión (∪):** Todos los elementos que están en al menos uno de los conjuntos.
  - Notación: x ∈ A ∪ B → x ∈ A ∨ x ∈ B

- **Complemento (A^c):** Elementos que no están en el conjunto.
  - Notación: x ∈ A^c → x ∉ A

### Ejemplo Gráfico:
- Si Ω es el conjunto universal y A, B, C son subconjuntos:
  - La intersección A ∩ B ∩ C sería la región común a todos.

## 3. Generalización de Operaciones:

### Familias de Conjuntos:
- **Conjuntos indexados:** A₁, A₂, …, Aₙ

### Intersección Generalizada:
- **Notación:** ⋂ᵢ=₁ⁿ Aᵢ
- **Definición:** {x ∈ Ω | ∀i ∈ {1, 2, …, n}, x ∈ Aᵢ}

### Ejemplo Formal:
- Intersección de una familia de conjuntos: {x ∈ Ω | ∀i ∈ {1, 2, …, n}, x ∈ Aᵢ}
